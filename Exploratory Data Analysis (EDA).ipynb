```python
# Q1 — DATA READINESS & QUALITY CHECKS
# ------------------------
print("===== Q1: Data Readiness & Quality Checks =====")
print("Shape:", df.shape)
print("\nColumn Types:\n", df.dtypes)
print("\nMissing Values per Column:\n", df.isna().sum())
print("\nNumber of Duplicates:", df.duplicated().sum())

# Handle missing numeric values (fill with median) & drop duplicates
numeric_cols = ['Study_Hours_per_Week','Attendance_Rate','Previous_Score','Quiz1','Quiz2','Quiz3','Final_Score']
for col in numeric_cols:
    if df[col].isna().sum() > 0:
        df[col].fillna(df[col].median(), inplace=True)

df.drop_duplicates(inplace=True)
df_cleaned = df

# Diagnostics Table
diagnostics = pd.DataFrame({
    'Column': df_cleaned.columns,
    'Type': df_cleaned.dtypes,
    'Missing Values': df_cleaned.isna().sum()
})
print("\nDiagnostics Table:\n", diagnostics)

print("\nDecisions Taken:")
print("- Filled missing numeric values with median to retain records.")
print("- Dropped duplicate rows to avoid double-counting.")
print("- Kept categorical columns as-is (no missing values).")
print("- Ready for descriptive statistics and plotting.\n")
```

```python
# ------------------------
# Q2 — DESCRIPTIVE STATISTICS & DISTRIBUTIONS
# ------------------------
print("===== Q2: Descriptive Statistics & Distributions =====")
desc_stats = df_cleaned[numeric_cols].describe().T
print("\nDescriptive Statistics:\n", desc_stats)

# Histograms / boxplots
plt.figure(figsize=(15,4))

plt.subplot(1,3,1)
plt.hist(df_cleaned['Final_Score'], bins=15, color='skyblue', edgecolor='black')
plt.title('Final Score Distribution')

plt.subplot(1,3,2)
plt.hist(df_cleaned['Study_Hours_per_Week'], bins=15, color='salmon', edgecolor='black')
plt.title('Study Hours per Week Distribution')

plt.subplot(1,3,3)
plt.hist(df_cleaned['Attendance_Rate'], bins=15, color='lightgreen', edgecolor='black')
plt.title('Attendance Rate Distribution')

plt.tight_layout()
plt.show()

print("\nInsights:")
print("- Final Score distribution is slightly right-skewed with some low outliers.")
print("- Study Hours show a few high-study outliers, mostly clustered 5-12 hours/week.")
print("- Attendance is mostly high (0.8-1.0), few low attendance outliers.")
print("- These distributions indicate most students are regular attendees but study hours and scores vary.")
```

```python
# ------------------------
# INSIGHTS
# ------------------------
print("\nInsights:")
print("- Final_Score distribution is roughly symmetric with some lower outliers.")
print("- Study_Hours_per_Week shows slight right skew; some students study more than average.")
print("- Attendance_Rate is mostly high (0.8–1.0) with few low values.")
print("- Outliers may indicate students needing additional support.")
print("- Numeric data is ready for correlation and deeper analysis.")
```

```python
# ------------------------
#  — CORRELATIONS & CROSS-TABS
# ------------------------
print("===== Q3: Correlation & Distributions =====")

# Correlation matrix
corr_matrix = df_cleaned[numeric_cols].corr()
print("\nCorrelation Matrix:\n", corr_matrix)

# Scatterplots
plt.figure(figsize=(15,4))

plt.subplot(1,2,1)
plt.scatter(df_cleaned['Study_Hours_per_Week'], df_cleaned['Final_Score'], color='skyblue', edgecolor='black')
plt.title('Study Hours vs Final Score')
plt.xlabel('Study_Hours_per_Week')
plt.ylabel('Final_Score')

plt.subplot(1,2,2)
plt.scatter(df_cleaned['Attendance_Rate'], df_cleaned['Final_Score'], color='salmon', edgecolor='black')
plt.title('Attendance Rate vs Final Score')
plt.xlabel('Attendance_Rate')
plt.ylabel('Final_Score')

plt.tight_layout()
plt.show()
```

```python
# Cross-tab: Average Final Score by Subject and Gender
avg_scores = df_cleaned.groupby(['Subject','Gender'])['Final_Score'].mean().reset_index()
print("\nAverage Final Score by Subject & Gender:\n", avg_scores)

print("\nInsights: Maths and Science students tend to have slightly higher scores with more study hours and attendance. Gender differences are minor but can guide interventions.")
```
